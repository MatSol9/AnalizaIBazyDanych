```python
import matplotlib.pyplot as plt
import numpy as np
f = lambda x : x**2 + 5
plt.plot(np.linspace(-1, 1), [f(x) for x in np.linspace(-1, 1) ])
plt.xlabel("x")
plt.ylabel("y")
plt.show()
plt.plot(np.linspace(-6, 6), [f(x) for x in np.linspace(-6, 6) ])
plt.xlabel("x")
plt.ylabel("y")
plt.show()
plt.plot(np.linspace(0, 5), [f(x) for x in np.linspace(0, 5) ])
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```
